```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
from skimage.color import rgb2gray
plt.rcParams['figure.figsize'] = (20, 6)
```

# Read image in and convert to grayscale

```python
image = rgb2gray(imread('cat.jpg'))
plt.imshow(image)
plt.axis('off')
plt.show()
```

# Perform SVD on data 

```python
U, S, VT = np.linalg.svd(image, full_matrices=False)
S_diag = np.diag(S)
```

# Compute approximate image 

```python
for r in (1, 5, 10, 25, 50):
    image_approx = U[:, :r] @ S_diag[:r, :r] @ VT[:r, :]
    plt.imshow(image_approx)
    plt.axis('off')
    plt.title('Rank = ' + str(r))
    plt.show()
```

# Using only rank 50 we were able to make out the original image almost completely. What proportion of the original energy is retained for rank 50?

```python
energy = np.sum(S[:50]) / np.sum(S)
print(energy)
```
